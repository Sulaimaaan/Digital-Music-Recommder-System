* The above two types of recommendations are not proper, because some cases where rating will be high but number of ratings will be poor
* So we need to calculate popularity score based on avg rating and number of ratings, 
    - for this data we can take **weighted average rating for each item as popularity score**
    
    Taking formula: 
    
               = ri * ni / avg(n)
               
               ri - rating of item
               ni - number of ratings of item
               n - number of ratings

```python
n = pdata['Rating_number'].mean()
```

```python
round(n,2)
```

```python
pdata['PopularityScore'] = round((pdata['Avg_rating']*pdata['Rating_number'])/n,2)
```

**Define & Training Model**

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# Model define cosine distance

cosinemodel = NearestNeighbors(n_neighbors=6, metric='cosine') 

# Number of neighbors can be adjusted according to our recommendation
# 6 neighbors mean including item taken with 5 neighbors.
```

```python
# Fit on Data

cosinemodel.fit(cdata)
```

```python
**Similar Recommendations for Item Taken**
```

```python
# Function for Prediciton

def recommend_items(itemtitle, n=5): # n is number of recommendations
    print("***********************************************")
    print(green("Content Based Recommendations:",['bold']))
    print("***********************************************")
    print()
    
    # Finding the item index
    idx = music[music['Title'].str.contains(itemtitle, case=False, na=False)].index
    if len(idx) == 0:
        return "Item not found!"
    
    idx = idx[0]  # Take first match if multiple exist
    
    
    print(red("Selected Item:",['bold']), music['Title'][idx])
    print()
    display(Image(url=music['Images'][idx]['large'], width=300, height=300))
    print()
    print(magenta("Store:",['bold']), music['Store'][idx])
    print()
    print(blue("Price: $ {}".format(music['Price'][idx]),['bold']))
    print()
    print("==========================================================")
    
    # Model input
    data = cdata.iloc[idx]
    
    # Find N nearest neighbors
    distances, indices = cosinemodel.kneighbors([data], n_neighbors=n+1)
    
    # distances are taken from cosine metric, these can be nearer to 0 for neighbors
    # indices are the indexes of items nearer

    # Get recommended item titles (excluding the input item)
    item_indices = indices.flatten()[1:]  # Skip first entry (input item)
    
    print(red("Recommended Top5 Items:",['bold']))
    print()
    for indx in item_indices[1:]:
        print(red("Title:",['bold']), music['Title'][indx])
        print("------------------------------------------------------")
        display(Image(url=music['Images'][indx]['large'], width=300, height=300))
        print()
        print(magenta("Store:",['bold']), music['Store'][indx])
        print()
        print(blue("Price: $ {}".format(music['Price'][indx]),['bold']))
        print()
```

```python
rui
```

* from the above rui , we need to predict missed(na) rating values.
* to work on this we need to know how users are giving ratings to items
* this information will be given by laten factors, which are calculated from rui matrix

                    rui = userlatent matrix * latent factors * item(transpose)latent matrix
                    
* this can be imlemented in python using a module **surprise**

**Define & fit**

```python
# pip install scikit-surprise 
# (if error)
# conda install -c conda-forge scikit-surprise

# Loading module and functions
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
# Data Reading to Surprise
reader = Reader(rating_scale=(0.5, 5.0))  # Ratings range from 0.5 to 5.0
data = Dataset.load_from_df(interactions[['UserID', 'ItemID', 'Rating']], reader)
```

```python
# Train-Test Split
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# Apply SVD
svd = SVD()
svd.fit(trainset)
```

```python
# Evaluate the model
predictions = svd.test(testset)
rmse = accuracy.rmse(predictions)
print(f"Root Mean Squared Error: {rmse:.4f}")
```

**Music Collaborative Recommendations:**

```python
# Creating  lookup dictionary
item_lookup = music.set_index("Music_id")["Title"].to_dict()
item_lookup
```

```python
# Compute item latent vectors
item_factors = {trainset.to_raw_iid(itemid): svd.qi[itemid] for itemid in trainset.all_items()}
```

```python
# Function to recommend similar items
def recommend_similar_music(item_title, n=5):
    
    item_id = music[music["Title"].str.contains(item_title, case=False, na=False)]["Music_id"]
    if item_id.empty:
        return []

    item_id = item_id.iloc[0]
    if item_id not in item_factors:
        return []

    item_vector = item_factors[item_id]
    # Compute cosine similarity with all other items
    similarities = {other_id: np.dot(item_vector, other_vector)
                    for other_id, other_vector in item_factors.items()
                    if other_id != item_id}

    # Get top N most similar items
    top_item_ids = sorted(similarities, key=similarities.get, reverse=True)[:n]
    
    return [(item_lookup[itemid], round(similarities[itemid],2)) for itemid in top_item_ids if itemid in item_lookup]
```

**User Collaborative - Item Recommendations**

```python
# Generate item recommendations for a specific user

def recommend_items(user_id, n=5):
    all_item_ids = interactions['ItemID'].unique()
    rated_items = interactions[interactions['UserID'] == user_id]['ItemID'].values
    notrated_items = [item for item in all_item_ids if item not in rated_items]
    
    predictions = [(item, svd.predict(user_id, item).est) for item in notrated_items]
    recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:n]
    
    return recommendations

# Example: Recommend 5 movies for user 1
recommended_items = recommend_items(user_id='ae25ivzx733cgiqslfydkv3j52pq', n=5)
print("Top 5 Recommended Items for User ae25ivzx733cgiqslfydkv3j52pq:")
print()
print((recommended_items))
```
